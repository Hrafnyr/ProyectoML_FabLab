```python
import pandas as pd

# Leer el archivo CSV (ajusta el nombre de archivo según el tuyo)
df = pd.read_csv('4DatasetListo.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Definir un subconjunto (ejemplo: primeras 20 columnas)
subset = df.columns[:20]
corr_subset = df.loc[:, subset].corr()

fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(
    corr_subset,
    cmap=sns.diverging_palette(150, 10, as_cmap=True),
    annot=True,
    fmt=".2f",
    ax=ax,
    square=True,
    cbar_kws={"shrink": 0.8}
)
ax.set_yticklabels(corr_subset.index, rotation=0)
plt.tight_layout()
plt.show()
```

## Interpretaciones

El mapa de calor es una forma visual para encontrar correlación entre variables mediante el color. En este cuadernillo se definió un color verde para la baja correlación; entre más oscuro el verde **menos** correlación lineal se tiene.

Entre más intenso sea el color rojo **más** correlación lineal se tiene.

* Terapia y TrataPsi: 0.51:

* Correlación por arriba de 0.3 entre
  variables del test PHQ9

  * Interpretación:
    > Los ítems del PHQ-9 están moderadamente correlacionados, lo que indica consistencia interna.
    > Es normal que quien reporta un síntoma depresivo (ej. anhedonia) también reporte otros relacionados (ej. fatiga, tristeza).
    > Este patrón es esperable en escalas unidimensionales, donde los ítems miden el mismo constructo (depresión).

* Las **correlaciones más altas** se presentan **dentro de cada test**, como era esperado.

### **Observaciones relevantes**

1. **Correlación interna en CD-RISC 10**

   * **Rango de correlación:** 0.3–0.5
   * **Interpretación:**

     * Los ítems del test de **resiliencia** presentan **consistencia interna moderada**.
     * Sugiere que las personas que puntúan alto en un aspecto de resiliencia tienden a hacerlo en los demás.

2. **Correlación interna en MHC 14**

   * **Rango de correlación:** 0.3–0.5
   * **Interpretación:**

     * Los ítems del **bienestar mental** muestran relación interna moderada.
     * Es esperable porque el MHC-14 evalúa tres dimensiones (emocional, social y psicológica) que están relacionadas pero no son idénticas.

3. **Correlación entre CD-RISC y MHC (0.3–0.4)**

   * **Interpretación:**

     * Existe una **relación positiva moderada**: a mayor resiliencia, mayor bienestar percibido.
     * Esto **confirma hallazgos teóricos** en psicología positiva: la resiliencia favorece el bienestar mental.
     * No es una correlación alta porque se trata de **constructos distintos**, aunque relacionados.

---

### **Conclusión preliminar del Cuadro 2**

* Tanto **CD-RISC** como **MHC** presentan **buena consistencia interna**.
* La correlación moderada entre ambos respalda su **complementariedad para análisis de clasificación y clustering**:

  * **Resiliencia** podría ayudar a diferenciar perfiles con igual depresión pero distinto bienestar.
  * **MHC** resume la percepción subjetiva de florecimiento psicológico.

## Modelos potenciales

**Clasificación:**

Se puede usar variables que no estén altamente correlacionadas para mejorar el poder predictivo.

Ej: Edad + SUMPHQ + SUMCDrisc + Sexo pueden tener bajo corr mutuo pero aportar a la predicción de mhc_dx.


**Clustering:**

Es ideal que no todas las variables estén muy correlacionadas, porque los clusters buscan patrones multidimensionales.

Si solo se toman variables redundantes, el clustering termina reflejando “lo mismo” que PHQ o GAD.
